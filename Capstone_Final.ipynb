```python
import pandas as pd
import tensorflow as tf
import numpy as np
from google.colab import drive
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Input, Embedding, Concatenate, Dense, Flatten
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.callbacks import EarlyStopping
```

```python
df_cleaned = df_cleaned.drop_duplicates()
df_cleaned.duplicated().sum()
```

```python
label_encoder = LabelEncoder()
df_cleaned['title_encoded'] = label_encoder.fit_transform(df_cleaned['title'])
df_cleaned
```

```python
scaler = MinMaxScaler()
columns_to_normalize = ['latitude', 'longitude']
df_cleaned[columns_to_normalize] = scaler.fit_transform(df_cleaned[columns_to_normalize])
df_title = df['title']
df_cleaned
```

```python
df_cleaned.drop('title', axis=1, inplace=True)
df_cleaned.dtypes
```

```python
df_cleaned.dtypes
```

```python
# Hasil Cleaning Data
df_cleaned = df_cleaned.drop_duplicates(subset=['title_encoded'])
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

```python


# Convert numerical columns to numeric type
numerical_cols = ['review_rating', 'latitude', 'longitude', 'Cuci saja', 'Dryclean', 'Komplit', 'Lainnya', 'Setrika saja']
df_cleaned[numerical_cols] = df_cleaned[numerical_cols].apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values after conversion
df_cleaned = df_cleaned.dropna(subset=numerical_cols)



# Split the data into training and testing sets
# Prepare input data
X_train_numerical = df_cleaned[numerical_cols].values
y_train = df_cleaned['title_encoded'].values

# Define model input layers
input_numerical = Input(shape=(len(numerical_cols),))

# Concatenate all features
all_features = input_numerical

# Add dense layers
all_features = Dense(128, activation='relu')(all_features)
all_features = Dense(64, activation='relu')(all_features)

# Output layer
output_layer = Dense(len(df_cleaned['title_encoded'].unique()), activation='softmax')(all_features)
# Create and compile the model
model = tf.keras.Model(inputs=input_numerical, outputs=output_layer)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_numerical, y_train, epochs=1500,batch_size = 32)
```

```python
print(model.summary())
```

```python
new_data_point = np.array([4.3, 0.200251, 0.995883, 1, 0, 1, 1, 1])
new_data_point = new_data_point.reshape(1, -1)
prediction = model.predict(new_data_point)
```

```python
predicted_class = np.argmax(prediction)
print("Predicted Class:", predicted_class)
```

```python
class_probabilities = prediction[0]

sorted_indices = np.argsort(class_probabilities)[::-1]

next_most_likely_indices = sorted_indices[1:11]

print("Next 10 Most Likely Classes:")
print(next_most_likely_indices)
```

```python
predicted_titles_encoded = next_most_likely_indices
predicted_titles = label_encoder.inverse_transform(predicted_titles_encoded)

print("Predicted Titles:")
for title in predicted_titles:
    print(title)
```

```python
filtered_rows = df_cleaned[df_cleaned['title_encoded'].isin(next_most_likely_indices)]
```
