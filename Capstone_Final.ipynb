##Model Development
Brief <br>
Model yang kami buat disini adalah model yang akan memprediksi id tempat laundry berdasarkan fitur longitude, latitude serta 5 kategori, nantinya model akan memberikan id prediksi dari fitur yang diberikan. Kemudian dari hasil prediksi tersebut kami ambil id dibawahnya yang memiliki karakteristik yang hampir sama dari id yang diprediksi.

```python
#Menggunakan semua dataset untuk training karena semua id diperlukan untuk diprediksi
X_train_numerical = df_cleaned[numerical_cols].values
y_train = df_cleaned['id'].values

input_numerical = Input(shape=(len(numerical_cols),))

all_features = input_numerical


all_features = Dense(512, activation='relu')(all_features)
all_features = Dense(128, activation='relu')(all_features)
all_features = Dense(64, activation='relu')(all_features)


output_layer = Dense((len(df_cleaned['id'].unique())+1), activation='softmax')(all_features)

model = tf.keras.Model(inputs=input_numerical, outputs=output_layer)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


model.fit(X_train_numerical, y_train, epochs=5000,batch_size = 32)
```
